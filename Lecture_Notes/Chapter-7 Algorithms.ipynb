Understanding the Efficiency Growth

In engineering software development, Big-O helps us understand tradeoffs between execution speed and memory consumption:

- O(1) (Constant Time): Execution time remains the same regardless of n, like accessing a specific array index.

- O(logn) (Logarithmic Time): Growth is very slow; often found in binary search algorithms.

- O(n) (Linear Time): Time grows in direct proportion to n, such as a single loop through sensor readings.

- O(n2) (Quadratic Time): Time grows exponentially with the input; common in nested loops or brute-force collision detection.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_big_o_complexity():
    """
    Plots the growth of different Big-O complexity classes 
    to demonstrate algorithm efficiency.
    """
    # Define input size (n)
    n = np.linspace(1, 20, 100)

    # Define Complexity Classes
    o_1 = np.ones(len(n))                  # Constant: O(1)
    o_log_n = np.log(n)                    # Logarithmic: O(log n)
    o_n = n                                # Linear: O(n)
    o_n_log_n = n * np.log(n)              # Linearithmic: O(n log n)
    o_n_2 = n**2                           # Quadratic: O(n^2)
    
    # Setup the plot
    plt.figure(figsize=(10, 8))
    plt.ylim(0, 50) # Limit y-axis to keep plot readable

    # Plot each line
    plt.plot(n, o_1, label='$O(1)$ - Constant', linewidth=2)
    plt.plot(n, o_log_n, label='$O(\log n)$ - Logarithmic', linewidth=2)
    plt.plot(n, o_n, label='$O(n)$ - Linear', linewidth=2)
    plt.plot(n, o_n_log_n, label='$O(n \log n)$ - Linearithmic', linewidth=2)
    plt.plot(n, o_n_2, label='$O(n^2)$ - Quadratic', linewidth=2)

    # Formatting according to professional engineering standards
    plt.title("Big-O Notation: Algorithm Efficiency Comparison", fontsize=14)
    plt.xlabel("Input Size ($n$)", fontsize=12)
    plt.ylabel("Operations / Execution Time", fontsize=12)
    plt.legend(loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Display the plot
    plt.show()

if __name__ == "__main__":
    plot_big_o_complexity()
```

In mechatronics and software engineering, understanding how your code scales is vital for maintaining high-frequency control loops.

Below are Python functions demonstrating common Big-O complexities using typical engineering data scenarios, such as reading sensor values or processing logs.

### 1. O(1) - Constant Time

The execution time is independent of the input size n. This is ideal for real-time control where you need to access a specific memory address or variable instantly.

```python
def read_latest_sensor(sensor_list):
    """
    O(1) Example: Accessing a specific index.
    No matter how many readings are in the list, 
    accessing one index takes the same amount of time.
    """
    return sensor_list[-1]  # Get the most recent reading

# Example usage
readings = [23.1, 23.5, 24.0, 23.8]
print(f"Latest reading: {readings[-1]}")
```

### 2. O(n) - Linear Time

The time to complete the task grows in direct proportion to the number of items. In mechatronics, this often occurs when you must process a stream of data to find a specific condition or calculate an average.

```python
def find_threshold_breach(data_stream, threshold):
    """
    O(n) Example: A single loop.
    In the worst case, we must check every single item.
    """
    for value in data_stream:
        if value > threshold:
            return f"Breach detected at {value}"
    return "All systems normal"

# Example usage
stream = [1.2, 1.5, 3.1, 0.8] # 'n' elements
print(find_threshold_breach(stream, 3.0))
```

### 3. O(n2) - Quadratic Time

The performance is proportional to the square of the input size. This usually results from nested loops. For example, comparing every robot's position against every other robot's position for collision detection is an O(n2) operation.

```python
def check_all_collisions(robot_positions):
    """
    O(n^2) Example: Nested loops.
    For 'n' robots, we compare each one to 'n' others.
    """
    collisions = []
    n = len(robot_positions)
    for i in range(n):
        for j in range(i + 1, n):
            # Calculate distance between robot i and robot j
            dist = abs(robot_positions[i] - robot_positions[j])
            if dist < 0.1:
                collisions.append((i, j))
    return collisions

# Example usage
robots = [0.5, 1.2, 0.55, 2.0]
print(f"Colliding pairs: {check_all_collisions(robots)}")
```
